***Importing and exporting data***

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randint(0, 5, (10, 5)))
df.to_csv("foo.csv")
```

```python
pd.read_csv("foo.csv")
```

```python
df.to_parquet("foo.parquet")
```

```python
pd.read_parquet("foo.parquet")
```

```python
df.to_excel("foo.xlsx", sheet_name="Sheet1")
```

```python
pd.read_excel("foo.xlsx", "Sheet1", index_col=None, na_values=["NA"])
```

“foo.xlsx”: Okunacak Excel dosyasının adı veya yolu.

“Sheet1”: Okunacak sayfanın adı (sayfa adı veya sayfa indeksi).

index_col=None: İndeks sütununu belirler. Eğer None olarak ayarlanırsa, indeks sütunu kullanılmaz.

na_values=[“NA”]: Sayısal olmayan sütunlarda yer alan “NA” değerlerini NaN (eksik veri) olarak işaretler.
