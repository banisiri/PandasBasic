```python
import pandas as pd
import numpy as np
```

```python
arrays = [
   ["bar", "bar", "baz", "baz", "foo", "foo", "qux", "qux"],
   ["one", "two", "one", "two", "one", "two", "one", "two"],
]

index=pd.MultiIndex.from_arrays(arrays, names=["first","second"])

df = pd.DataFrame(np.random.rand(8,2), index=index, columns=["A","B"])

df2 = df[:4]

df2
```

```python
stacked = df2.stack(future_stack=True)
stacked
```

```python
stacked.unstack()
```

```python
stacked.unstack(1)
```

```python
stacked.unstack(0)
```

**Pivot tables**

```python
df = pd.DataFrame(
    {
        "A": ["one", "one", "two", "three"] * 3,
        "B": ["A", "B", "C"] * 4,
        "C": ["foo", "foo", "foo", "bar", "bar", "bar"] * 2,
        "D": np.random.randn(12),
        "E": np.random.randn(12),
    }
)
df
```

pandas.pivot_table() fonksiyonunun temel parametreleri şunlardır:

data: İşlem yapılacak DataFrame.

values: İşlem yapılacak sütun veya sütunlar (sayısal veriler).

index: Pivot tablosunun satır indeksini oluşturan sütun veya sütunlar.

columns: Pivot tablosunun sütun indeksini oluşturan sütun veya sütunlar.

aggfunc: İşlem türü (örneğin, ‘mean’, ‘sum’, ‘count’ vb.).

Diğer parametreler: fill_value, margins, dropna, vb.

```python
pd.pivot_table(df, values="D", index=["A","B"], columns=["C"])
```
