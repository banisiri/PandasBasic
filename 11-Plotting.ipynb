```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
plt.close("all")
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range("14.04.2024", periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
df = pd.DataFrame(
    np.random.randn(1000,4), index=ts.index, columns=["A","B","C","D"]
)

df = df.cumsum()
plt.figure()
df.plot()
plt.legend(loc="best")
```
